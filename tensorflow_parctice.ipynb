```python
#the equation is x**2+7x+12
#first we intialize the unknown value as 0 
unknown=tf.Variable(0.0)
a=tf.constant(7.0)
b=tf.constant(12.0)
c=tf.multiply(unknown,7)
d=tf.pow(unknown,2)
e=tf.add(c,d)
equation=tf.add(e,b)
zerovalue=tf.constant(0.0)
diff=tf.square(equation-zerovalue)
tf.compat.v1.disable_eager_execution()
solving=tf.compat.v1.train.GradientDescentOptimizer(.001)
with tf.compat.v1.Session() as sess:
  tf.compat.v1.global_variables_initializer()
  for i in range(1000):
    sess.run(solving)
  print(sess.run(unknown))

    




```

```python
from tensorflow import keras

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as se 
import numpy as np 
```

```python
raw_dataset=pd.read_csv('/content/auto-mpg.csv',na_values='?')
dataset=raw_dataset.copy()
dataset.tail()
#calculate the frequency of a df column value 
frequency_df=dataset['car name'].value_counts()
print(frequency_df)
```

```python
dataset.isna().sum()
#ok then , there is no nan values 
```

```python
origin=dataset.pop('origin')
```

```python
dataset['USA']=(origin==1)*1.0
dataset['Europe']=(origin==2)*1.0
dataset['Japan']=(origin==3)*1.0
dataset.tail()
```

```python
#train test split 
traindata=dataset.sample(frac=.8,random_state=0)
testdata=dataset.drop(traindata.index)
```

```python
traindata.head()
```

```python
se.pairplot(traindata[['mpg','cylinders','displacement','weight']],diag_kind='kde')
```

```python
train_stats=dataset.describe()
train_stats.pop('mpg')
train_stats=train_stats.transpose()
train_stats
```

```python
train_label=traindata.pop('mpg')
test_label=testdata.pop('mpg')
```

```python
normed_traindata=(traindata-train_stats['mean'])/train_stats['std']
normed_testdata=(testdata-train_stats['mean'])/train_stats['std']
normed_testdata.dropna(inplace=True,axis=1)
normed_traindata.dropna(inplace=True,axis=1)
#ttt.fillna(value='', inplace=True)

normed_testdata.tail()
normed_traindata.tail()
```

```python
model=keras.Sequential()
model.add(keras.layers.Dense(64,activation=keras.activations.relu,input_shape=[len(normed_traindata.keys())]))
model.add(keras.layers.Dense(64,activation=keras.activations.relu))
model.add(keras.layers.Dense(1))

model.compile(loss='mean_squared_error',optimizer=keras.optimizers.RMSprop(.001),metrics=['mean_absolute_error','mean_squared_error'])

```

```python
history=model.fit(normed_traindata,train_label,epochs=1000,validation_split=.2,verbose=0,)
```

```python
hist=pd.DataFrame(data=history.history)
hist.head()
```

```python
hist['epochs']=history.epoch
hist.head()
```

```python
se.pairplot(hist[['epochs','loss','mean_squared_error','mean_absolute_error']])
```

```python
plt.figure()
plt.xlabel(xlabel='epochs')
plt.ylabel(ylabel='mean_absolute_error')
plt.plot(hist['epochs'],hist['mean_absolute_error'],label='train error')

plt.xlabel(xlabel='epochs')
plt.plot(hist['epochs'],hist['val_mean_absolute_error'],label='val error')
plt.ylim([0,5])
plt.legend()

plt.figure()
plt.xlabel(xlabel='epochs')
plt.ylabel(ylabel='Mean Squared error')
plt.plot(hist['epochs'],hist['mean_squared_error'],label='train error')
plt.plot(hist['epochs'],hist['val_mean_absolute_error'],label='val error')
plt.ylim([0,20])
plt.legend()
plt.show()
```

```python
directory='/content/PetImages'
categories=['dog','cat']
data=[]

for categorey in categories:
  path=os.path.join(directory,categorey)
  classNum=categories.index(categorey)
  for img in tqdm(os.listdir(path)):
    try:
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        resizedimage=cv2.resize(img_array,(80,80))
        data.append([resizedimage,classNum])
    except Exception as e : 
      pass    
    
  
   
    
      



```

```python
import random
random.shuffle(data)
sample=data[:1]
print(sample[0])
for sample in data[:10]:
    print(sample[1])
```

```python
xdata=[]
ydata=[]
for data,labels in data:
  xdata.append(data)
  ydata.append(labels)
```

```python
print(xdata[0].reshape(-1,80,80,1))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xdata,ydata,test_size=.2,random_state=40)
#convert the data into numpy array so the the nn could manipulates it 
xtrain=np.array(xtrain)
xtest=np.array(xtest)
ytrain=np.array(ytrain)
ytest=np.array(ytest)
```

```python
model=keras.Sequential()
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128,activation=keras.activations.sigmoid))
model.add(keras.layers.Dense(128,activation=keras.activations.sigmoid))
model.add(keras.layers.Dense(2,activation=keras.activations.softmax))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(xtrain,ytrain,epochs=3)
```

```python
testloss,testaccuracy=model.evaluate(xtest,ytest)
#that is good , no overfitting 
```
